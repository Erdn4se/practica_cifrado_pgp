## Prueba de cifrado asimetrico PGP

### 1. Importacion de librerias

```python
import pgpy
import os
```

### 2. Importamos clave publica y mensaje

```python
# 0 .Definimos las rutas de los ficheros de la llave pública y del mensaje a cifrar
ruta_llave = os.path.join('llavero', 'publicKey.txt')
ruta_mensaje = os.path.join('mensaje', 'mensaje_2.txt')

# 1. Leemos la llave publica
with open(ruta_llave, 'r') as f:
    key_data = f.read()
public_key, _ = pgpy.PGPKey.from_blob(key_data)

# 2. Leemos el mensaje a cifrar respetando acentos y caracteres especiales (encoding='utf-8')
with open(ruta_mensaje, 'r', encoding='utf-8') as f:
    contenido_mensaje = f.read()

print(f"Llave pública cargada: {public_key.userids[0].name}")
print(key_data)
print("Mensaje original leído correctamente:")
print('#####################################')
print(contenido_mensaje)
```

### 3. Encriptamos mensaje mediante clave publica

```python
# Ciframos el mensaje utilizando la llave publica
mensaje = pgpy.PGPMessage.new(contenido_mensaje)
mensaje_cifrado = public_key.encrypt(mensaje)

bloque_pgp_final = str(mensaje_cifrado)
print("--- RESULTADO DEL CIFRADO PGP ---")
print(bloque_pgp_final)
```

### 4. Generamos par de clave publica y privada utilizando GPG

```python
private_key_2 = pgpy.PGPKey.new(pgpy.constants.PubKeyAlgorithm.RSAEncryptOrSign, 2048)
uid = pgpy.PGPUID.new('Usuario de Prueba', email='usuario.prueba@example.com')
private_key_2.add_uid(uid)
public_key_2 = private_key_2.pubkey
print("Par de claves generado:")
print(f"Usuario: {uid.name}, Email: {uid.email}")
print(f"Clave pública: {public_key_2}")
print(f"Clave privada: {private_key_2}")
```
